```python
def draw_boxes(image, boxes, confidences, class_ids, labels=None, colors=None):
    height, width, _ = image.shape
    if labels is None:
        labels = [str(i) for i in range(len(boxes))]
    if colors is None:
        colors = [(0, 255, 0) for _ in range(len(boxes))]

    for i, (x, y, w, h) in enumerate(boxes):
        x1, y1, x2, y2 = int(x - w / 2), int(y - h / 2), int(x + w / 2), int(y + h / 2)
        cv2.rectangle(image, (x1, y1), (x2, y2), colors[class_ids[i]], 2)
        label = "{}: {:.2f}".format(labels[class_ids[i]], confidences[i])
        text_size = cv2.getTextSize(label, cv2.FONT_HERSHEY_SIMPLEX, 0.5, 2)[0]
        cv2.rectangle(image, (x1, y1 - text_size[1] - 4), (x1 + text_size[0], y1), colors[class_ids[i]], -1)
cv2.putText(image, label, (x1, y1 - 2), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2)
```

```python
draw_boxes(image,result,2,1)
```
